```python
data=pd.read_csv('data/berkeley.csv')
```

```python
data
```

```python
x=array(data['YEAR'])
y=array(data['metANN'])
y[y==999.9]=nan
```

```python
x
```

```python
y
```

```python
station='BERKELEY'

plot(x,y,'-o')
ylabel('Temperature [C]')
title(station)
```

```python
model=ols('y ~ x', data={'y':y,'x':x})
results=model.fit()
results.summary()
```

```python
xx=linspace(min(x)-10,max(x)+10,20)
yy=results.predict({'x':xx})

m=results.params['x']
mÏƒ=results.bse['x']

b=results.params['Intercept']

plot(x,y,'-o')
ylabel('Temperature [C]')
title(station+" : y=%.3g (+/- %.4g) x + %.3g" % (m,mÏƒ,b))
plot(xx,yy,'b-')
```
